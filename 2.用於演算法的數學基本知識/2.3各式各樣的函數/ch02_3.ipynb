## 問題 2.3.3 ★
請畫出下列函數的圖形。  
1. $y = 2x + 3$
2. $y = 10^x$
3. $y = \log_4 x$
4. $y = \frac{\log_2 x}{2}$

```python
import matplotlib.pyplot as plt
import numpy as np


# 1.畫圖的範圍
x_linear = np.linspace(-2, 5, 500)

# 對數函數的定義域必須 > 0，我們從 0.01 開始取點
x_log = np.linspace(0.01, 5, 500)
# 定義 四個函數
y1 = 2 * x_linear + 3
y2 = 10**x_linear
y3 = np.log(x_log) / np.log(4)  # 換底公式
y4 = np.log2(x_log) / 2


# 畫法一：四個函數合併在同一張圖 (Combined Plot)
# ==========================================
plt.figure(figsize=(10, 6))

plt.plot(x_linear, y1, "r-", label=r"$y_1 = 2x + 3$")
plt.plot(x_linear, y2, "g-", label=r"$y_2 = 10^x$")
# 因為 y3 和 y4 完全相同，y4 使用虛線(---)疊加，方便觀察重合
plt.plot(x_log, y3, "b-", linewidth=2.5, label=r"$y_3 = \log_4 x$")
plt.plot(x_log, y4, "m--", linewidth=1.5, label=r"$y_4 = \frac{\log_2 x}{2}$")

# 設定座標軸核心（通過原點的十字線）與網格
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, linestyle=":", alpha=0.6)

# 關鍵：限制顯示範圍，否則 10^5 會讓圖表失去焦點
plt.xlim(-1, 5)
plt.ylim(-3, 15)

plt.title("Combined Functions Plot", fontsize=14)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig("combined_functions_plot.png", dpi=300)



# 畫法二：用 2x2 子圖分開畫 (Subplots)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 子圖 1: y = 2x + 3
axs[0, 0].plot(x_linear, y1, "r-", label=r"$y = 2x + 3$")
axs[0, 0].set_title(r"$y = 2x + 3$")

# 子圖 2: y = 10^x
axs[0, 1].plot(x_linear, y2, "g-", label=r"$y = 10^x$")
axs[0, 1].set_ylim(-1, 100)  # 稍微限制一下 Y 軸看清趨勢
axs[0, 1].set_title(r"$y = 10^x$")

# 子圖 3: y = log_4(x)
axs[1, 0].plot(x_log, y3, "b-", label=r"$y = \log_4 x$")
axs[1, 0].set_title(r"$y = \log_4 x$")

# 子圖 4: y = log_2(x) / 2
axs[1, 1].plot(x_log, y4, "m-", label=r"$y = \frac{\log_2 x}{2}$")
axs[1, 1].set_title(r"$y = \frac{\log_2 x}{2}$")

# 為所有子圖加上軸線與網格
for ax in axs.flat:
    ax.axhline(0, color="black", linewidth=0.5)
    ax.axvline(0, color="black", linewidth=0.5)
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.legend()

plt.tight_layout()
plt.savefig("functions_plot.png", dpi=300)

```

## 問題 2.3.6 ★★
地震的規模是以對數表示地震能量大小的值。已知規模每增加 $1$，能量會變為約 $32$ 倍（準確地說是 $\sqrt{1000}$ 倍）。  
在此，假設規模每增加 $1$，能量正好變為 $32$ 倍，請回答以下問題。  
1. 規模 6.0 地震的能量是規模 5.0 地震的幾倍？  
2. 規模 7.3 地震的能量是規模 5.3 地震的幾倍？  
3. 規模 9.0 地震的能量是規模 7.2 地震的幾倍？

```python
# 根據題目意，可以反推其地震能量(x)與地震規模(y)的函數關係:
# y = (1/5)*log_{2} (x)
import math

def power(x):
    return (1/5)*math.log(x)

def inverse_p(y):
    return 2**(5*y)

# 1.
print(inverse_p(6-5))
# 2.
print(inverse_p(7.3-5.3))
# 3.
print(inverse_p(9-7.2))

```

## 關於地震模型的補充：
我建立的模型，是僅由數學條件「規模每增加 $1$，能量會變為約 $32$ 倍」，在不考慮「真正的芮氏規模 $0$ 代表多少能量（基準點）」的  
$$ y = \frac{1}{5} \log_2 (x)$$
現實中，地震能量 $E$（焦耳）與地震規模 $M$ 的標準關係式（Gutenberg-Richter 關係式）通常寫作：
$$\log_{10} E = 4.8 + 1.5M$$

## 問題 2.3.8 ★★★★
在 desmos.com 上繪製各種圖形，找出一個函數 $y = f(x)$ 滿足以下所有條件。
1. 對於所有實數 $a$，$0 < f(a) < 1$。
2. $f(x)$ 是單調遞增的函數。

## 畫圖
有以下：  

![四圖合一](function.png)



1. 反正切函數改     
$$f(x)\ =\frac{1}{\pi}\arctan\left(x\right)+\frac{1}{2}$$


2. 邏輯斯函數(Logistic Function / Sigmoid) : 機器學習常用到  
$$f_1(x)=\frac{1}{1+e^{-x}}$$
特點： 當 $x \to \infty$ 時，$e^{-x} \to 0$，分母趨近於 1，函數值趨近於 1；  
當 $x \to -\infty$ 時，$e^{-x} \to \infty$，分母變得極大，函數值趨近於 0。  
它的圖形是一個非常平滑的 S 型曲線。  

3. 雙曲正切修正函數 (Hyperbolic Tangent Variant) 
$$f_2(x) = \frac{1}{2}\tanh(x) + \frac{1}{2}$$
與 第一個類似，利用雙曲正切函數 $\tanh(x)$ 來修改。  

4. 代數函數 (Algebraic Variant)
$$f_3(x) = \frac{1}{2} \left( \frac{x}{\sqrt{1 + x^2}} \right) + \frac{1}{2}$$

  

更一般化的版本:
其實只要 $g(x)$ 它滿足嚴格單調遞增且值域為 $(A, B)$，  
這個公式：$$f(x) = \frac{g(x) - A}{B - A}$$
就是符合題意的，可試著證明：
1. 值域的變換：  
因為 $A < g(x) < B$，我們同時減去 $A$，得到：$$0 < g(x) - A < B - A$$
因為 $B > A$，所以 $B - A > 0$。我們同時除以 $(B - A)$，不等號方向不變：$$0 < \frac{g(x) - A}{B - A} < 1$$  
這就證明了 $0 < f(x) < 1$。  

2. 單調遞增性的保持：  
假設 $x_1 < x_2$，因為 $g(x)$ 是單調遞增的，所以 $g(x_1) < g(x_2)$。  
當我們對不等式兩邊同時減去一個常數 $A$，再除以一個正數 $(B - A)$ 時，不等號的方向完全不會改變。  
因此必然有 $f(x_1) < f(x_2)$，這證明了 $f(x)$ 依然是單調遞增。


如果今天給定的 $g(x)$ 值域是半有界或全無界的，也可以透過簡單的複合函數把它拉回 $(0, 1)$：

如果值域是 $(0, \infty)$（例如 $g(x) = e^x$）：  
我們可以直接用分式變換：$f(x) = \frac{g(x)}{1 + g(x)}$。  
這樣當 $g(x) \to \infty$ 時，$f(x) \to 1$；當 $g(x) \to 0$ 時，$f(x) \to 0$。  
如果值域是 $(-\infty, \infty)$：  
這就是最開始想到的 $\arctan(x)$，或是利用 $e^{g(x)}$ 先把值域壞轉成 $(0, \infty)$，再套用上面的方法。
