## 大 $O$ 符號（Big O Notation）的嚴謹數學定義

設 $f(N)$ 與 $g(N)$ 為定義在正實數或正整數上的兩個函數。
若存在兩個正常數（Positive Constants）$c$ 與 $N_0$，使得：

$$\forall N \ge N_0, \exists c > 0, \quad 0 \le f(N) \le c \cdot g(N)$$

則我們可將此關係記作：

$$f(N) = O(g(N))$$

---

### 🔍 核心概念解析

1. **集合的概念（最嚴謹的視角）**：
   在高等數學中，$O(g(N))$ 其實是一個**函數集合**。
   $$O(g(N)) = \{ f(N) \mid \exists c > 0, N_0 > 0 \text{ 使得 } \forall N \ge N_0, 0 \le f(N) \le c \cdot g(N) \}$$
   因此，寫成 $f(N) = O(g(N))$ 是一種習慣用法，更嚴謹的數學表示法其實是 **$f(N) \in O(g(N))$**。

2. **漸近上界（Asymptotic Upper Bound）**：
   這個定義的意思是，當 $N$ 成長到足夠大（超過 $N_0$）之後，函數 $f(N)$ 的增長速度將會被常數倍的 $g(N)$ 壓在下方。大 $O$ 符號描述的是演算法在「最壞情況（Worst Case）」下的時間複雜度趨勢。

請證明 $\log_2 N$ 和 $\log_{10} N$ 之間最多只有常數倍的差異。  
這也是蘭道大 $O$ 符號中以省略 $O(\log N)$ 和 $\log$ 底數的形式書寫的原因。

## 說明：使用換底公式:

$\log_{10} N  = \frac{\log_2 {N}}{\log_2 10} = \frac{1}{\log_2 10} \cdot \log_2 N$

兩者只差常數倍而已，根據時間複雜度的嚴謹公式，兩者是一樣的

## 2.4.4 解答

| 計算次數 | 執行時間基準 | $N \log_2 N$ | $N^2$ | $2^N$ |
| :--- | :--- | :--- | :--- | :--- |
| $10^6$ 次以內 | 0.001 秒以下 | $N \le 60,000$ | $N \le 1,000$ | $N \le 20$ |
| $10^7$ 次以內 | 0.01 秒以下 | $N \le 520,000$ | $N \le 3,162$ | $N \le 23$ |
| $10^8$ 次以內 | 0.1 秒以下 | $N \le 4,500,000$ | $N \le 10,000$ | $N \le 26$ |
| $10^9$ 次以內 | 1 秒以下 | $N \le 40,000,000$ | $N \le 31,622$ | $N \le 29$ |

說明：
很單純的計算而已。  
當計算次數為 $10^9$ 時：$N^2 \le 10^9 \implies N \le \sqrt{10^9} \approx 31,622$。
(這就是為什麼平常寫題時，如果資料量 $N$ 在 $10^4$ 到 $3 \times 10^4$ 之間，$O(N^2)$ 的演算法通常還勉強能過，一跨過這界線就會超時。)
