## 問題 2.2.1 ★
1萬、1億、1兆分別是 10 的幾次方？

```python
import math

print(int(math.log(10000,10)))
print(int(math.log(100000000,10)))
print(int(math.log10(1000000000000)))

# 注意：若是 print(math.log(1000000000000,10)) 將會顯示 11.999999999999998
# 這是精度的問題，看下面的補充
# math.log10(x) 內部有針對底數 10 進行優化，會比math.log(x, 10) 更好

```

## 精確度的問題：為何 print(math.log(1000000000000,10)) 將會顯示 11.999999999999998 ?

Python 的 math.log(x, 10) 在底層通常是透過換底公式實作的，  
類似於：$$\frac{\ln(1000000000000)}{\ln(10)}$$  
電腦背後是直接將兩個數值做運算，電腦背後並無無理樹的概念，所以當數值越大，會有精度的問題

# 心得： 位元運算
運算過程如下：
1. 先將數字轉為二進位
2. 各個位數進行邏輯運算
3. 再將數字轉回十進位數字

實際驗算：  
I.  

1.
$$
13_{10} \ \& \ 14_{10}= 1101_2 \ \& \ 1110_2= 1100_2= 12_{10}
$$  
2.  
$$
13_{10} \ \ | \ 14_{10}= 1101_2 \ \ | \ 1110_2= 1111_2= 15_{10}
$$
3.  
$$
13_{10} \oplus 14_{10}= 1101_2 \oplus 1110_2= 0011_2= 3_{10}
$$

II.  
$$
8_{10} \oplus 4_{10} \oplus 2_{10}\oplus 1_{10}= 1000_2 \oplus 100_2 \oplus 10_{2} \oplus 1_{2} = 1111_2= 15_{10}
$$


## 問題 2.2.4 問題 ID：005 ★★  
給定 $N$ 個整數 $a_1, a_2, a_3, \dots, a_N$。  
請製作輸出 $(a_1 + a_2 + a_3 + \dots + a_N) \pmod{100}$ 值的程式。  

```python
# 輸入一個list 版本
def modten(number):
    return sum(number) %100

print(modten([100,200,34]))
```

```python
# 使用者輸入n : 代表n個數字要相加，再輸入n個數字

def modten1(n):
    n = int(input("請輸入 N: "))
    total = 0
    for i in range(n):
        total = total + int(input())
    return total %100

print(f"結果: {modten1(n)}")
```
