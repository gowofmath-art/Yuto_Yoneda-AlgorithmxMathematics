## 問題 2.1.4 ★  
請以手算解出以下問題：  

請將 1001（二進制表示）轉換為十進制。  

請將 127（十進制表示）轉換為二進制、三進制。  

```python
def twototen(n):
    x = 0
    count = 0
    while n > 0:
        a = n % 10
        n = n // 10
        x = x + a * (2**count)
        count = count +1

    return x  


def tentotwo(n):
    x = 0
    count = 0
    while n > 0:
        a = n % 2
        n = n // 2
        x = x + a* (10**count)
        count = count +1 
    return  x

def tentothree(n):
    x = 0
    count = 0
    while n > 0:
        a = n % 3
        n = n // 3
        x = x + a* (10**count)
        count = count +1 
    return  x

print(twototen(1001))
print(tentotwo(127))
print(tentothree(127))
```
