### 問題 3.7.1 ⭐
請以逐個計算的方法，求出滿足以下遞迴式的數列之**第 1 項至第 10 項**：

1. $a_1 = 1, \quad a_n = 2a_{n-1} \quad (n \ge 2)$
2. $a_1 = 1, \quad a_2 = 1, \quad a_3 = 1, \quad a_n = a_{n-1} + a_{n-2} + a_{n-3} \quad (n \ge 4)$

````python
# 問題3.7.1
# 遞迴寫法：
def question_371(n):
    if n == 1:
        return 1
    return 2*question_371(n-1)

n = 10
for i in range(1, n + 1):
    print(question_371(i))

# 一般項寫法：這就是等比級數

def question_371_1(n):
    return 2**(n-1)

n = 10
for i in range(1, n + 1):
    print(question_371_1(i))

def question_371_2(n):
    return [2**(i-1)  for i in range(1,n+1)]    
print(question_371_2(10))
````

### 問題 3.7.2 ⭐⭐
在下列的道路網中，從「起點」到「終點」的**最短路徑**有幾種方法？

```mermaid
graph LR
    Start([起點]) --> Cross1[交叉路口 1]
    Cross1 --> Cross2[交叉路口 2]
    Cross2 --> Cross3[交叉路口 3]
    Cross3 --> Cross4[交叉路口 4]
    Cross4 --> End([終點])

    %% 上方跨越箭頭
    Start([起點]) --> Cross2
    Cross1 --> Cross4
    Cross3 --> End
```


## 說明：
定義狀態：設 dp[X] 代表「從起點走到 X 節點的最短路徑方法數」。  
尋找依賴關係（狀態轉移）：要到達某個路口，只要看「有誰可以直接連到它」，並把那些來源的方法數相加。  
dp[起點] = 1  
dp[路口1]：只能從起點過來 $\rightarrow$ dp[起點]  
dp[路口2]：只能從路口1過來 $\rightarrow$ dp[起點] + dp[路口1]  
dp[路口3]：可以從路口1或路口2過來 $\rightarrow$ dp[路口2]  
dp[路口4]：可以從路口2或路口3過來 $\rightarrow$ dp[路口1] + dp[路口3]  
dp[終點]：可以從路口3或路口4過來 $\rightarrow$ dp[路口3] + dp[路口4]  


因此最短路徑其實只有兩條：  
1.起點>交叉路口1>交叉路口4>終點  
2.起點>交叉路口2>交叉路口3>終點

### 練習一：青蛙的移動一般化

* **給定參數**：
  * 棚架總數為 $N$，其高度陣列為 $h = [h_1, h_2, \dots, h_N]$。
  * 限制條件：$2 \le N \le 100000$。
* **行動與代價規則**：
  * 從棚架 $i$ 跳到 $i+1$：消耗體力 $|h_i - h_{i+1}| \quad (1 \le i \le N-1)$
  * 從棚架 $i$ 跳到 $i+2$：消耗體力 $|h_i - h_{i+2}| \quad (1 \le i \le N-2)$
* **求解目標**：從棚架 1 移動到棚架 $N$ 的最小總體力消耗。


````python
# 輸入 H 是一個高度list ，N = len(H)
# 設：從第一個移動到第i個棚架時消耗的最小體力為dp[i]

# 時間複雜度為 O(n)

def frog_move(H):
    n = len(H)
    dp = [0]*n
    
    dp[0] = 0
    dp[1] = abs(H[0]-H[1])
    for i in range(2,n):
        v1 = dp[i-1] + abs(H[i-1] - H[i])
        v2 = dp[i-2] + abs(H[i-2] - H[i])
        print(v1 , v2)
        dp[i] = min(v1 , v2)
    return dp

print(frog_move([8,6,9,2,1]))

````

### 2.4.6 計算次數例④：全搜尋與指數時間 (問題 ID: 009)

#### 📝 問題描述
從左數來第 $i \quad (1 \le i \le N)$ 張卡片上寫了整數 $A_i$。有沒有辦法從卡片中選擇幾張，使總和正好成為 $S$ 呢？

* **範例輸入**：
  * $N = 3$
  * $S = 11$
  * $(A_1, A_2, A_3) = (2, 5, 9)$
* **範例輸出**：
  * **Yes** （因為選擇卡片 1 和卡片 3 的話，總和為 $2 + 9 = 11$）

#### ⚠️ 限制條件（Constraints）
* $1 \le N \le 60$
* $1 \le A_i \le 10000$
* $1 \le S \le 10000$
* **執行時間限制**：1 秒



## 解法說明：

全搜尋一定不可能，如果$N=60$，全部的可能性來到 $2^60$ 就算給電腦跑也會超時。  
此題必須使用本章節提到的動態規劃法。

有兩個方法：  
第一個是將時間複雜度降下來的「二維陣列法」，第二個則是進一步把空間省下來的「一維陣列滾動優化法」。

````python
def solve_subset_sum(A: list, S: int) -> bool:
    """
    判斷是否能從卡片列表 A 中選擇幾張，使總和正好成為 S。
    """
    N = len(A)
    
    # 建立 (N + 1) * (S + 1) 的二維布林表格，預設全部為 False
    dp = [[False] * (S + 1) for i in range(N + 1)]
    
    # 陣列的初始化（邊界條件）：
    # 考慮 0 張卡片時，唯有目標總和為 0 的狀態為 True（什麼都不選）
    dp[0][0] = True
    
    # 動態規劃主迴圈：i 代表考慮前 i 張卡片，j 代表當前目標總和
    for i in range(1, N + 1):
        card_value = A[i - 1]  # 目前這張卡片上的整數值
        for j in range(0, S + 1):
            
            # 情況 A：j 小於當前卡片值，無法選擇這張卡片，直接繼承上一列的結果
            if j < card_value:
                dp[i][j] = dp[i - 1][j]
                
            # 情況 B：j >= card_value，可以選擇方法 A（不選）或方法 B（選擇）
            else:
                # 只要「不選這張卡片能湊出 j」或「選了這張卡片後，剩餘的額度能被湊出」其中一個成立即可
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - card_value]
                
    # 最終答案：考慮完全部 N 張卡片後，能不能湊出目標總和 S
    return dp[N][S]

# --- 測試範例（書中的範例輸入） ---
if __name__ == "__main__":
    cards = [2, 5, 9]
    target = 11
    
    result = solve_subset_sum(cards, target)
    print(f"有沒有辦法湊出 {target}？ 答案是: {'Yes' if result else 'No'}")
````
