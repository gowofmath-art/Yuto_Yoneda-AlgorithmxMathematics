    %% 上方跨越箭頭

## 說明：
定義狀態：設 dp[X] 代表「從起點走到 X 節點的最短路徑方法數」。  
尋找依賴關係（狀態轉移）：要到達某個路口，只要看「有誰可以直接連到它」，並把那些來源的方法數相加。  
dp[起點] = 1  
dp[路口1]：只能從起點過來 $\rightarrow$ dp[起點]  
dp[路口2]：只能從路口1過來 $\rightarrow$ dp[起點] + dp[路口1]  
dp[路口3]：可以路口2過來 $\rightarrow$ dp[路口2]  
dp[路口4]：可以從路口2或路口3過來 $\rightarrow$ dp[路口1] + dp[路口3]  
dp[終點]：可以從路口3或路口4過來 $\rightarrow$ dp[路口3] + dp[路口4]  


因此最短路徑其實只有兩條：  
1.起點>交叉路口1>交叉路口4>終點  
2.起點>交叉路口2>交叉路口3>終點

### 練習一：青蛙的移動一般化

* **給定參數**：
  * 棚架總數為 $N$，其高度陣列為 $h = [h_1, h_2, \dots, h_N]$。
  * 限制條件：$2 \le N \le 100000$。
* **行動與代價規則**：
  * 從棚架 $i$ 跳到 $i+1$：消耗體力 $|h_i - h_{i+1}| \quad (1 \le i \le N-1)$
  * 從棚架 $i$ 跳到 $i+2$：消耗體力 $|h_i - h_{i+2}| \quad (1 \le i \le N-2)$
* **求解目標**：從棚架 1 移動到棚架 $N$ 的最小總體力消耗。


````python
# 輸入 H 是一個高度list ，N = len(H)
# 設：從第一個移動到第i個棚架時消耗的最小體力為dp[i]

# 時間複雜度為 O(n)

def frog_move(H):
    n = len(H)
    dp = [0]*n
    
    dp[0] = 0
    dp[1] = abs(H[0]-H[1])
    for i in range(2,n):
        v1 = dp[i-1] + abs(H[i-1] - H[i])
        v2 = dp[i-2] + abs(H[i-2] - H[i])
        print(v1 , v2)
        dp[i] = min(v1 , v2)
    return dp

print(frog_move([8,6,9,2,1]))

````

## 練習二：上樓梯的方法

### 問題描述：
* **題目**：爬上 $N$ 階樓梯，一步可選擇爬 **1 階** 或 **2 階**，求從第 0 階到達第 $N$ 階的方法數。  
* **限制條件**：$1 \le N \le 45$ ，時間限制：1秒  


````python
# 此為費氏數列，但還是必須練習
# 時間複雜度:O(n)、空間複雜度：O(n)  (因為建立了list)

def stairs(N):
    dp = [0]*(N+1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2,N+1):
        dp[i] = dp[i-1] + dp[i-2]
        print(dp)
    return  dp    

print(stairs(10))

````

````python
# 優化：空間複雜度來到O(1)
# 這是AI幫忙的！  未來還需要手刻

def stairs_optimized(N):
    if N == 0 or N == 1:
        return 1
    
    # 不開 list，只用兩個變數代表 dp[i-2] 和 dp[i-1]
    prev2 = 1  # 相當於 dp[0]
    prev1 = 1  # 相當於 dp[1]
    
    for i in range(2, N + 1):
        # 當前這一階，就是前兩步相加
        current = prev1 + prev2
        
        # 滾動更新：把狀態往後傳，準備給下一圈迴圈用
        prev2 = prev1
        prev1 = current
        
    return current  # 最終只回傳最後那個滾動出來的目標值

print(stairs_optimized(10))  # 一樣會是 89！
````

## 練習三：背包問題

### 問題描述：
* **問題情境**：
  太郎有 $N$ 個物品要放進一個最大容量（耐重）為 $W$ 的背包中。
* **已知條件**：
  每個物品 $i \quad (1 \le i \le N)$ 都具備兩個屬性：
  * **重量**：$w_i$
  * **價值**：$v_i$
* **限制與目標**：
  每個物品只能選擇**「放（1）」**或**「不放（0）」**，且放入背包的物品**總重量不能超過 $W$**。請在不超重的前提下，求出背包中物品**總價值的最大值**。
  * **限制條件**：$1 \le N \le 100$ ，$1 \le W \le 10^5$，$1 \le w_i \le W$ ， $1 \le v_i \le 10^9$
  * **時間限制：2秒  

````python
# 在前面的範例中，最難的背包問題，務必要自己手刻！
# 參數說明：w : 耐重、wi：物品i的重量 (list)、vi：物品i的價值(list)
# 建立二維陣列：dp[i][j]：只考慮前 i 個物品，在總重量不超過 j 時的最大價值
# 說穿了，就是依照順序去選取，對每一個物品、每一個容量上限，分別比較「不選這個物品」與「選這個物品」兩種情況，儲存較大的價值。

# 時間與空間複雜度：O(NW)  
# 建立了 (N+1)*(W+1) 的表格

def backpack(W,w,v):
    n = len(w)
    dp = [[0 for i in range(W+1)] for j in range(n+1) ]


    for i in range(1,n+1):
        for j in range(W+1):
            if j < w[i-1]:
                dp[i][j] = dp[i-1][j]
            if j >= w[i-1]:
                dp[i][j] = max(dp[i-1][j] , dp[i-1][j-w[i-1]] + v[i-1])
    print(dp)
    answer = 0
    for i in range(W+1):
        answer = max(answer , dp[n][i])


    return answer

# 範例
print(backpack(10,[3,6,4,2] , [100,210,130,57]))
````

## 解法說明：

全搜尋一定不可能，如果$N=60$，全部的可能性來到 $2^{60} $ 就算給電腦跑也會超時。  
此題必須使用本章節提到的動態規劃法。

有兩個方法：  
第一個是將時間複雜度降下來的「二維陣列法」，第二個則是進一步把空間省下來的「一維陣列滾動優化法」。

````python
# 這是AI生的，未來還需多手刻！

def solve_subset_sum(A: list, S: int) -> bool:
    """
    判斷是否能從卡片列表 A 中選擇幾張，使總和正好成為 S。
    """
    N = len(A)
    
    # 建立 (N + 1) * (S + 1) 的二維布林表格，預設全部為 False
    dp = [[False] * (S + 1) for i in range(N + 1)]
    
    # 陣列的初始化（邊界條件）：
    # 考慮 0 張卡片時，唯有目標總和為 0 的狀態為 True（什麼都不選）
    dp[0][0] = True
    
    # 動態規劃主迴圈：i 代表考慮前 i 張卡片，j 代表當前目標總和
    for i in range(1, N + 1):
        card_value = A[i - 1]  # 目前這張卡片上的整數值
        for j in range(0, S + 1):
            
            # 情況 A：j 小於當前卡片值，無法選擇這張卡片，直接繼承上一列的結果
            if j < card_value:
                dp[i][j] = dp[i - 1][j]
                
            # 情況 B：j >= card_value，可以選擇方法 A（不選）或方法 B（選擇）
            else:
                # 只要「不選這張卡片能湊出 j」或「選了這張卡片後，剩餘的額度能被湊出」其中一個成立即可
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - card_value]
                
    # 最終答案：考慮完全部 N 張卡片後，能不能湊出目標總和 S
    return dp[N][S]

# --- 測試範例（書中的範例輸入） ---
if __name__ == "__main__":
    cards = [2, 5, 9]
    target = 11
    
    result = solve_subset_sum(cards, target)
    print(f"有沒有辦法湊出 {target}？ 答案是: {'Yes' if result else 'No'}")
````

問題 3.7.6 (問題 ID: 031) ⭐⭐⭐  
太郎的暑假有 $N$ 天，已知在第 $i$ 天（$1 \le i \le N$）學習的話，會提升 $A_i$ 的實力。但是他不想連續學習兩天。  
太郎在暑假期間實力最多會提高多少？請製作一個程式求其最大值。

````python
# A是一個 每天學習多少的陣列
# 善用動態規劃法！
# dp[i] : 第i天的最大學習經驗  
# 不想連續學習兩天：可跳一天(-2天)或是跳兩天(-3天)  (沒有跳三天的情況)

def learn(A):
    n = len(A)
    if n == 0:
        return 0
    if n == 1:
        return A[0]
    if n == 2:
        return max(A[0] , A[1])


    dp = [0]*(n)

    dp[0] = A[0]
    dp[1] = max(A[0] , A[1])
    dp[2] = max(A[0] + A[2] , A[1])

    for i in range(3,n):
        # 三種情況！ 1.隔一天 2.隔2天 3.不動
        m1 = dp[i-2] + A[i]
        m2 = dp[i-3] + A[i]
        m3 = dp[i-1]
        print(m1 ,m2, m3)
        dp[i] = max(m1 , m2, m3)
        print(dp)

    # 如果想用 $n$ 來拿最後一格：要寫 dp[n-1]，直接寫dp[-1] 代表最後一個！
    return dp[-1]

print(learn([1,2,1,2,5]))


````
