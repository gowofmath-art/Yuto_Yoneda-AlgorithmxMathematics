## 說明：

## 問題 3.4.2 ★  
請計算當參加依照以下機率獲得獎金的賭注時，獲得獎金的期望值。  
另外，參加費為 500 日圓時，請判斷參加是有利還是損失。  

| | 1 等獎 | 2 等獎 | 3 等獎 | 4 等獎 | 5 等獎 |
| :--- | :---: | :---: | :---: | :---: | :---: |
| **獎金** | 100 萬 | 10 萬 | 1 萬 | 1000 | 0 |
| **機率** | $\frac{1}{10000}$ | $\frac{9}{10000}$ | $\frac{9}{1000}$ | $\frac{9}{100}$ | $\frac{9}{10}$ |


```python
# Expected Value期望值：
E = 1000000*(1/1000)+100000*(9/10000)+10000*(9/1000)+1000*(9/100)+0*(9/10)
print(E)

def Profit(n):
    if E - n > 0:
        return "賺錢"
    else:
            return "賠錢"
print(Profit(500))
 
```

## 問題 3.4.3 問題 ID：025 ★★★  

次郎的暑假有 $N$ 天。他按照以下順序來決定第 $i$ 天（$1 \le i \le N$）的學習時間。  
🔵 在一天的開始擲骰子  
🔵 擲了骰子出現 1、2 時：學習 $A_i$ 時間  
🔵 擲了骰子出現 3、4、5、6 時：學習 $B_i$ 時間  
請製作一個程式，求出他的暑假總學習時間的期望值。  

```python
# 假設 A , B是兩個list，為第i天的學習時間，N = len(A) = len(B)

def learn(A,B):
    n = len(A)
    a = 0
    b = 0
    for i in range(n):
        a = a + A[i]/3
    for i in range(n):
        b = b + 2*B[i]/3   
    return a + b
```

## 說明：
每天都會有 $\frac{1}{3}$ 機率是 學習 $A_i$ 時間，會有 $\frac{2}{3}$ 機率是 學習 $B_i$ 時間。

## 問題 3.4.4 問題 ID：026 ★★★★  
有一臺機器，如果支付 1 美元，就會從 $N$ 種硬幣中以等機率出現 1 個。  
請製作一個程式，計算在收集到所有類型的硬幣前所支付的金額的期望值（提示：➡️ 第 2.5.10 項）。



```python
def money_value(n):
    E = 0
    for i in range(1,n+1):
        E = E + 1/i

    total = n*E    
    return total
```

## 說明：  

先想不同情況，如果： 
第 1 階段：收集第 1 種硬幣身上沒有任何硬幣。  
投第 1 次進去，不論拿到哪一種，都一定是「新種類」。
成功機率是 $\frac{N}{N} = 1$。  
期望要投的次數：$1$ 次。  

第 2 階段：收集第 2 種硬幣手上已經有 1 種了，還剩 $N-1$ 種沒看過。
現在投 1 美元，拿到「新種類」的機率是 $\frac{N-1}{N}$。  
期望要投的次數（機率的倒數）：$\frac{N}{N-1}$ 次。  

第 3 階段：收集第 3 種硬幣手上已經有 2 種了，剩 $N-2$ 種沒看過。  
拿到新種類的機率是 $\frac{N-2}{N}$。  
期望要投的次數：$\frac{N}{N-2}$ 次。  

依此類推……直到：第 $N$ 階段：收集最後一種硬幣（大魔王）手上已經有 $N-1$ 種了，只差最後 1 種。  
投進去能拿到最後那種的機率只有 $\frac{1}{N}$。期望要投的次數：$\frac{N}{1} = N$ 次。  

數學公式總和根據期望值的線性性質（Linearity of Expectation），我們把每一階段的期望次數全部加起來，就是總花費的期望值（因為投一次是 1 美元，次數就等於金額）：  

$E = \frac{N}{N} + \frac{N}{N-1} + \frac{N}{N-2} + \dots + \frac{N}{1}$ 

把分子共有的 $N$ 提出來：  

$E = N \times \left( \frac{1}{1} + \frac{1}{2} + \frac{1}{3} + \dots + \frac{1}{N} \right)$  

括號裡面的那一串，就是數學上著名的調和級數（Harmonic Number），通常記作 $H_N$。

這也是著名的：彩券收集問題(Coupon collector's problem)
