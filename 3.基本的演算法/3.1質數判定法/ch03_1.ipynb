## 問題 3.1.1 ★  
請使用第 3.1.2 項所述的方法，判定自己的年齡是否為質數。

3.1.2 :快速的質數判定法，只判斷到整數的根號即可。 O(n^(1/2))


```python
# 順便複習一下，這是最單純的質數判定法，時間複雜度為O(n)
def is_prime(n):
    for i in range(2,n):
        if n%i == 0:
            return False
    return True

print (is_prime(15))
```

```python
# 這是快速的質數判定法。時間複雜度為O(n^(1/2))

def is_prime1(n):
    for i in range(2,int(n**(1/2))):
        if n%i == 0:
            return False
    return True    

print (is_prime1(17))
```

## 問題 3.1.2   問題 ID : 014 ★★  
請製作將自然數 $N$ 分解為質因數的程式。  
所謂質因數分解是指將自然數以質數相乘的形式來表示，如下：  
$286 = 2 \times 11 \times 13$    
$20211225 = 3 \times 5 \times 5 \times 31 \times 8693$  

計算複雜度最好為 $O(\sqrt{N})$。

```python
# 列舉因數 (時間複雜度為O(n^(1/2)))
def factor(n):
    fa = []
    for i in range(1,int(n**(1/2)+1)):
        if n%i == 0 :
            fa.append(i)
            if i != n//i:
                fa.append(n//i)
    return fa
print (factor(25))
```

```python
# 美化 (使得因數的順序一致)
def factor1(n):
    small = []
    big = []
    for i in range(1,int(n**(1/2)+1)):
        if n%i == 0 :
            small.append(i)
            if i != n//i:
                big.append(n//i)
    big.reverse()            
    return small + big
print (factor1(60))
```

```python
# 建立一個list 將這些數放進去
def prime_factorization(n):
    factor = []
    for i in range(2,int(n**(1/2)+1)):
        while n%i == 0 :
            factor.append(i)
            n = n//i
    if n > 1:        
        factor.append(n)
    return factor
print(prime_factorization(60))
```

```python
# 寫成 *號形式  (較為美觀)

def prime_factorization1(n):
    factor = []
    for i in range(2,int(n**(1/2)+1)):
        while n%i == 0 :
            factor.append(i)
            n = n//i
    if n > 1: 
        factor.append(n)

    factor1 = [str(x) for x in factor]
    factor2 = "*".join(factor1)

    return factor2
print(prime_factorization1(1020))

```

## 說明：質因數分解  

根據算術基本定理，自然數N的標準分解是必然唯一。  
想法也是利用數學思考，將一個正整數N做質因數分解，流程如下：  
1. 從小數字i(1,2,3...) 開始去除。  
2. 若除得盡 (if n%i == 0) 則將i(質因數)寫入 list (也就是N的質因數) 並且將n 更新為n//i。  
3. 如果n還可以被i整除，就要繼續除，直到不能整除為止，對應到程式碼中的while迴圈。  
4. 不斷重複1&2&3，也就是對應到程式碼中的for迴圈。  
5. 如果最後的n大於1，也必須加入list (因為可能有大質因數)， 如果是1就不用了。
