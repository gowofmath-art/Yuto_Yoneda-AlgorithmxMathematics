## 在開始解題之前，務必先實作：選擇排序！

演算法的步驟如下所示意。  
反覆進行 $N - 1$ 次以下操作。在第 $i$ 次操作中，進行以下事項。

1. 在未排序部分（從 $A_i$ 到 $A_N$）中尋找最小的元素 $A_{min}$
2. 將 $A_i$ 和 $A_{min}$ 交換

此時，操作後的 $A_1、A_2、\dots、A_N$ 依從小到大的順序排列。




````python
# A 是一個放了許多數字的list
def selection_sort(A):
    n = len(A)
    for i in range(n-1):  #不用n次，因為交換到最後一次時，最後一個必然是最大值
        min_index = i
        for j in range(i+1,n):  
        # 檢查i之後的那個陣列，這邊的for迴圈使要找後面的陣列中哪個元素最小，並且動態調整其索引，標記索引為j，為了是下一步讓A[i] 與 A[min_index] 互換
            if A[j] < A[min_index]:  #如果後面的A_j < A_min (選定的第一個該排序的元素)
                min_index = j        #將j賦值給 min_index，記錄目前最小元素的索引
                print(i,j)  # printt 出來觀察
        A[i] , A[min_index] = A[min_index] , A[i]  
        print (A)    # 可print出來觀察

    return A

# 測試看看
print(selection_sort([5,2,1,3,4,2]))  
````


### 實作階乘的遞迴函式

以下為遞迴函式的演算法步驟：  
**操作 $N$：**

* $N$ 為 1 時：回傳 1
* $N$ 為 2 以上時：回傳（操作 $N - 1$ 的計算結果）$\times N$



````python
def factorial (n):
    if n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(5))
````

## 實作以分割統治法求合計值

思考：求N個整數的合併值問題  (不要用for迴圈)
搭配遞迴函式。


````python
# A 是一個數字list，l , r是表示要從這個list的哪裡加到哪裡
# 這個函式使用的是左閉右開區間：[l,r)
# 在 Python 慣用的 [l,r) 表示法中，r 是終點邊界，不是要被取用的索引。

def divide_sum(A, l, r):
    if r-l == 1:
        return A[l]
    m = (l+r)//2
    s_1 = divide_sum(A, l, m)
    s_2 = divide_sum(A, m, r)
    return s_1 + s_2
    
A = [1,4,5,6,7,8]
answer = divide_sum(A, 1, 3)
print (answer)
````

## 問題 3.6.2 ★★★★★
請證明藉由以下演算法可以將長度為 $N$ 的陣列 $A[1]、A[2]、\dots、A[N]$ 依從小到大的順序排列。  
另外，本排序演算法發表於 2021 年 10 月。  

c ++ 語言
```cpp
for (int i = 1; i <= N; i++) {
    for (int j = 1; j <= N; j++) {
        if (A[i] < A[j]) swap(A[i], A[j]);
    }
}

````python
# 先用 python 複製一份

def x(A):
    n = len(A)
    for i in range(n):
        for j in range(n):
            if A[i] < A[j]:
                A[i] , A[j] = A[j] , A[i]
                print(i,j)  # print出來觀察
                print(A)    # print出來觀察
    return A

print (x([5,3,2,1,4]))
````


## 邏輯說明與複雜度分析

1. **時間複雜度**：
   使用了兩層巢狀 `for` 迴圈，外圈與內圈皆完整遍歷 $N$ 次，因此總比較次數為 $N \times N$，時間複雜度為 $O(N^2)$。

2. **運作機制**：
   此演算法並非比較索引，而是比較陣列中兩個位置的**數值大小**。當滿足條件 `A[i] < A[j]` 時，便立即觸發 `swap` 交換兩者。

3. **與傳統排序的差異**：
   雖然同為 $O(N^2)$，但它與「選擇排序」不同。選擇排序在內圈中只尋找極值索引、且每輪只交換一次；而本演算法的內圈每次都從 $1$ 跑到 $N$，在執行過程中會引發大量、看似混亂的頻繁交換，最終透過迴圈的對稱性讓陣列達到正確排序。

## Isort (或稱 4-Line Sort / 世界上最簡單的排序法)
故事背景：  
這段看似不合理卻能正確排序的程式，在 Stanley P. Y. Fung 於 2021 年提交的論文 Is this the simplest (and most surprising) sorting algorithm ever? 中受到分析。論文證明了演算法的正確性，並討論其看似反直覺的性質。   

從結構上看，它跟傳統排序一樣是兩層 $O(N^2)$ 的迴圈，而且都在做 swap。  

但它們在邏輯上的最大差別在於：選擇排序（Selection Sort）：是有強烈目的性的。它每一輪都「瞄準」未排序部分的最小值，找到後才精準交換 $1$ 次。  
Isort：則是利用數學的對稱性與收斂。它完全不去找最小值在哪，而是透過兩層硬性跑滿 $N \times N$ 的迴圈，讓大數與小數在 $i > j$ 的矩陣空間中不斷碰撞、自動洗牌，最後藉由對稱性「剛好」收斂成正確答案。

````python
# 完整的合併排序：
# A 是一個需要被排序的陣列

def merge_sort(A):
    n = len(A)
    if n <= 1 :
        return A
    # Divide：切成左右兩半
    m = len(A) // 2

    l = A[:m]
    r = A[m:]
    print(A)  # 觀察

    # Conquer：遞迴排序左右兩半
    sorted_left = merge_sort(l)
    sorted_right = merge_sort(r)

    # Combine：合併兩個已排序陣列
    return merge(sorted_left, sorted_right)

def merge(l,r):
    i = 0
    j = 0
    answer = []
    while ( i < len(l) and j < len(r)):
        if l[i] <= r[j]:
            answer.append(l[i])
            i = i+1
        else:
            answer.append(r[j])
            j = j+1
    while i < len(l):
        answer.append(l[i])
        i = i+1
    while j < len(r):
        answer.append(r[j])
        j = j+1
    return answer    


print (merge_sort([5,3,7,2,6,1,4,8]))
````
