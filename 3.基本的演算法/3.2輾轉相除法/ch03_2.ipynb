## 解：

| 步驟數 | 0 | 1 | 2 | 3 | 4 | 5 | 6 |
| :--- | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| $A$ 的值 | 372 | 372 | 104 | 104 | 14 | 14 | 0 |
| $B$ 的值 | 506 | 134 | 134 | 30 | 30 | 2  | 2 |

因此可以得知，最大公因數為2。


```python
# 使用 程式碼來實現
# 法1. 最單純的找最大公因數
# 就是從 1到min(A,B) 中所有的正整數去一一驗證，若有找到就更新

def great_common_divisor(A,B):
    gcd = 1
    for i in range(1,min(A,B)+1):
        if (A%i == 0) and (B%i == 0) :
            gcd = i
    return gcd
print (great_common_divisor(1012,506))

```

```python
# 法2. 使用輾轉相除法
# 使用wlile迴圈判斷 每一次A,B誰大誰小，較大者更新為 大%小，直到不能除為止。

def Euclidean_algorithm(A,B):
    while (A >= 1 and B >= 1) :
        if A < B :
            B = B%A
        else:
            A = A%B  
        print (A,B)
    return A if B == 0 else B
print (Euclidean_algorithm(180,300))

```

## 說明：為何輾轉相除法的時間複雜度為 $O(\log(A+B))$？

**核心定理：** 在輾轉相除法中，每一次操作後， $A+B$ 的值一定會減少到原本的 $\frac{2}{3}$ 以下。
> 註：以下只討論 $A > B$ 的情況（若 $A < B$，操作一次後兩數角色即對調，結論相同）。

設操作前的兩數為 $A, B$，操作一次後，較大的數 $A$ 會被替換為餘數 $R = A \pmod B$，新總和變為 $B + R$。
我們可以根據 $A$ 是否大於等於 $B$ 的 $2$ 倍，分為兩種情況討論：

### 1. 當 $A < 2B$ 時（差小於 2 倍，即 $A-B < B$）
此時 $A$ 只能除以 $B$ 一次，餘數為 $R = A - B$。
* **操作前的總和：** 因為 $A < 2B$，所以 $A + B < 3B$。
* **操作後的總和：** 新總和為 $B + R = B + (A - B) = A$。
* **縮小比例：**
  $$\frac{\text{操作後總和}}{\text{操作前總和}} = \frac{A}{A+B}$$
  因為 $A < 2B \implies 3A < 2A + 2B \implies 3A < 2(A+B)$，得知 $A < \frac{2}{3}(A+B)$。
  因此，操作後的總和 $A$ 確實小於原本總和的 $\frac{2}{3}$ 倍。

---

### 2. 當 $A \ge 2B$ 時（差大於等於 2 倍，即 $A-B \ge B$）
此時 $A$ 可以除以 $B$ 至少兩次，餘數 $R$ 必定嚴格小於 $B$（即 $R < B$）。
* **操作前的總和：** 因為 $A \ge 2B$，所以 $A + B \ge 3B \implies B \le \frac{1}{3}(A+B)$。
* **操作後的總和：** 新總和為 $B + R$。因為 $R < B$，所以 $B + R < 2B$。
* **縮小比例：**
  將 $B \le \frac{1}{3}(A+B)$ 代入上式：
  $$B + R < 2B \le 2 \cdot \frac{1}{3}(A+B) = \frac{2}{3}(A+B)$$
  因此，操作後的總和 $B+R$ 也確實小於原本總和的 $\frac{2}{3}$ 倍。

---

### ⏳ 複雜度總結
不論哪種情況，每執行一次操作，新的總和必定滿足：
$$(A+B)_{\text{new}} \le \frac{2}{3}(A+B)_{\text{old}}$$

因為 $A$ 和 $B$ 都是正整數，當演算法結束時，總和至少大於等於 $1$。若假設總共操作了 $L$ 次：
$$\left(\frac{2}{3}\right)^L \cdot (A+B) \ge 1$$

兩邊取對數（對數底數為 $\frac{3}{2}$）可得：
$$L \le \log_{\frac{3}{2}}(A+B)$$

由換底公式可知，這與以 $2$ 或 $e$ 為底的對數僅差一個常數倍，因此時間複雜度為：
$$O(\log(A+B))$$



## 問題 3.2.2 ❪ 問題 ID：016 ❫ ★★  
製作一個程式，使用輾轉相除法，計算 $N$ 個正整數 $A_1、A_2、\dots、A_N$ 的最大公因數。  

```python
# 先定義好函式：兩數求最大公因數
def gcd_two(a,b):
    if b == 0:
        return a
    return gcd_two(b, a % b)

# 利用上面的函式去求多個正整數的最大公因數，允許輸入一個list(放多個正整數)
def great_common_divisor2(number):
    n = len(number)
    if n == 1:
        return number[0]
    
    gcd = gcd_two(number[0],number[1])
    for i in range(2, n):
        gcd = gcd_two(gcd , number[i])
        print (gcd)
    return gcd    

print (great_common_divisor2([12,4,18,12]))
```

```python
# 業界的寫法：AI推薦的
def great_common_divisor3(number):
    gcd = number[0]
    
    # 2. 從第二個數字（索引 1）開始，逐一跟目前的 current_gcd 算新的 GCD
    for i in number[1:]:  # 這裡使用切片(Slice)寫法，代表從索引 1 拿到最後
        gcd = gcd_two(gcd, i)
        
    return gcd

# 測試看看
print(great_common_divisor3([24, 60, 36]))  # 預期輸出: 12
print(great_common_divisor3([7]))           # 預期輸出: 7

```

## 說明：多個正整數求最大公因數
如果有 $N$ 個數字，  
1. 先拿出前兩個數字，算出它們的 GCD。  
2. 拿這個「目前的 GCD」跟第三個數字算 GCD，得到新的「目前 GCD」。  
3. 拿新的「目前 GCD」跟第四個數字算 GCD...   
4. 一直重複這個過程，直到把列表中所有的數字都消耗完，最後留下來的就是這 $N$ 個數的總最大公因數！

## 問題 3.2.3 ❪ 問題 ID：017 ❫ ★★★  
製作一個程式，使用輾轉相除法，計算 $N$ 個正整數 $A_1、A_2、\dots、A_N$ 的最小公倍數。  
（來源：AOJNTL_1_C-Least Common Multiple）

```python
# 先定義好函式：輾轉相除法求最大公因數
def gcd_two(a,b):
    if b == 0:
        return a
    return gcd_two(b, a % b)

# 再定義一個函式：求兩數的最小公倍數
def lcm_two(a,b):
    c = gcd_two(a,b)
    return c*(a//c)*(b//c) 
print(lcm_two(36,60))

# 利用上述函式，求最小公倍數
def least_common_multiple(number):
    n = len(number)
    if n == 1:
        return number[0]
    
    lcm = lcm_two(number[0],number[1])
    for i in range(2, n):
        lcm = lcm_two(lcm , number[i])
        print (lcm)
    return lcm    
    

print (least_common_multiple([2,4,6,8]))
```

```python
# 優化板！

def least_common_multiple1(number):
    lcm = number[0]
    
    for i in number[1:]:  
        lcm = lcm_two(lcm, i)
        
    return lcm

# 測試看看
print(least_common_multiple1([24, 60, 36]))  
print(least_common_multiple1([7]))          


```
