### 問題 3.3.1 ★  
請分別計算 $_2C_1$、$_8C_5$、$_7P_2$、$_{10}P_3$ 的值。  

```python
# 試著用程式碼來寫！
# 此方法為先定義階乘函數，在透過排列與組合呼叫此階乘函數完成兩函數

def Factorial(n):
    if n == 0:
        return 1
    return n*Factorial(n-1) 
print (Factorial(5))
# 進行n次乘法  O(n) 

# 排列：
def permutation(n,r):
    #p(n,r)=n!//(n-r)!
    return Factorial(n)//Factorial(n-r)
# 進行 n+(n-r)次乘法  O(n)

print(permutation(5,3))

# 組合：
def combination(n,r):
    #c(n,r)=n!//(r!)*(n-r)! 
    return Factorial(n)//(Factorial(r)*Factorial(n-r))

print(combination(5,2))

# 進行 n+r+(n-r)次的乘法  O(n)
# 問題在於多做了很多不必要的運算。
```

```python
# 優化

def permutation_1(n,r):
    if r < 0 or r > n:
        return 0
    p = 1
    for i in range(n,n-r,-1):
        p = p * i
    return p
print(permutation_1(5,3))
# 只需要算r次即可，O(r)，會快很多

def combination_1(n,r):
    if r < 0 or r > n:
        return 0
    r = min(r,n-r)
    c = 1
    for i in range(1,r+1):
        c = c * (n-i+1)//i
    return c
print(combination_1(5,2))

# 只需要算r次即可，O(r)，會快很多
```

### 問題 ID：018

便利商店裡有 $N$ 個商品，第 $i$ 個（$1 \le i \le N$）商品的價格是 $A_i$ 日圓。購買 2 種不同商品的方法中，合計價格為 500 日圓的方法有幾種？

**條件：** $2 \le N \le 200000$，$A_i$ 為 100、200、300、400 中的任意一個

**執行時間限制：** 1 秒

```python
#預設A是一個商品價格的list，代表第1-N的商品價格，而N=len(A)

def id018 (A):
    a = 0
    b = 0
    c = 0
    d = 0
    for price in A:
        if price == 100:
            a += 1
        elif price == 200:
            b += 1
        elif price == 300:
            c += 1
        elif price == 400:
            d += 1            

    return  (a*d + b*c)
```

```python
# 全搜尋的爛方法： 
# 總共要算 C(n,2)次

def id018_1(A):
    count = 0
    for i in range(len(A)):
        for j in range(i,len(A)+1):
            if A[i] + A[j] == 500:
                count += 1
    return count

```

## 說明：  

直接找 100、200、300、400 各有幾個 (假設a,b,c,d個)  
500 = 100+400 = 200+300  
因此最後只需要 $a*d+ b*c$ 這麼多次

全搜尋一定會 超時 時間複雜度來到：$O(n^2)$

### 問題 ID：019  

有 $N$ 張卡片，左起第 $i$ 張（$1 \le i \le N$）卡片的顏色為 $A_i$。$A_i = 1$ 時為紅色，$A_i = 2$ 時為黃色，$A_i = 3$ 時為藍色。選擇 2 張相同顏色卡片的方法有幾種？  

**條件：** $2 \le N \le 500000$，$1 \le A_i \le 3$  

**執行時間限制：** 1 秒  


```python
# 預設A是一個顏色的list，代表第1-N的顏色，而N=len(A)
def id019(A):
    r = 0
    y = 0
    b = 0
    for i in A:
        if i == 1:
            r += 1
        elif i == 2:
            y += 1
        elif i == 3:
            b += 1   
             
    return r*(r-1)//2 + y*(y-1)//2 + b*(b-1)//2

```

```python
# 全搜尋的爛方法
# 總共要算 C(n,2)次

def id019_1(A):
    count = 0
    for i in range(len(A)):
        for j in range(i,len(A)+1):
            if A[i] == A[j] :
                count += 1

    return count 
```

## 說明:
全搜尋一定會 超時 時間複雜度來到：$O(n^2)$  

因此改用更好的做法，排列組合的觀點：
找紅黃藍三色分別有幾個：假設為分別為r,y,b  
因此：  
選擇 2 張相同顏色卡片的方法為 $_rC_2+  _yC_2 + _bC_2 $


### 問題 3.3.6  `問題 ID：022` ★★★  
有 $N$ 張卡片，左起第 $i$ 張卡片寫有整數 $A_i$。請製作一個程式，求出總和為 $100000$ 的 2 張卡片的選擇方法有幾種。   
在滿足 $2 \le N \le 200000$、$1 \le A_i \le 99999$ 的情況下，預期在 1 秒以內結束執行。

```python
## 全搜尋的爛方法
# 總共要算 C(n,2)次
def id022(A):
    n = len(A)
    count = 0
    for i in range(n):
        for j in range(i,n+1):
            if A[i] + A[j] == 100000:
                count += 1
    return count
```

```python
# 預設A是一個卡片的list，代表第1-N張卡片上的整數，而N=len(A)

def id022_array(A):
    n = len(A)
    total = 0
    # 創建一個0的list 
    count = [0]*100001
    
    # 統計每個數字出現了幾次，最後count這個list會呈現 每個整數出現了幾次
    for i in A:
        count[i] = count[i] +1

    # 成對出現，只要這樣做即可
    for i in range(1,50000):
        pair = 100000 -i 
        total = total + count[i] * count[pair]

    # 最後加上 整數為50000的情況
    total = total + count[50000]*(count[50000]-1)//2

    return total
```
